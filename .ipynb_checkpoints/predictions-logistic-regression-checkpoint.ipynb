```python
#query to get the total actors that have worked in each film

query='''SELECT f.film_id,COUNT(fa.actor_id) as no_of_actors FROM film f
LEFT JOIN film_actor fa
USING (film_id)
GROUP BY f.film_id'''

film_actors=pd.read_sql_query(query,engine)
film_actors
```

```python
#query to get the category of each film

query="""SELECT f.film_id,c.name as category FROM category c
JOIN film_category fc
USING (category_id)
JOIN film f
USING (film_id)
ORDER BY f.film_id;"""

category=pd.read_sql_query(query,engine)
category
```

```python
#getting copies of each movie in inventory
#used left join as there may be movies that are not present in the inventory and not to drop them

query="""SELECT f.film_id,COUNT(i.film_id) as film_inventory FROM film f 
LEFT JOIN inventory i
USING (film_id)
GROUP BY film_id;"""

film_inventory=pd.read_sql_query(query,engine)
film_inventory
```

```python
#query to get the total number of times a movie was rented in may

query="""SELECT f.film_id, COUNT(r.rental_id) as total_rented FROM rental r
JOIN inventory i
USING (inventory_id)
JOIN film f 
USING (film_id)
WHERE MONTH(r.rental_date)=05
GROUP BY f.film_id;"""

rental_counts_may=pd.read_sql_query(query,engine)
rental_counts_may
```

### 2. Create a query to get the list of all unique film titles and a boolean indicating if it was rented (rental_date) in May 2005. (Create new column called - 'rented_in_may'). This will be our TARGET (y) variable.

```python
query='''SELECT DISTINCT i.film_id as rented_in_may FROM rental r
JOIN payment p
ON r.customer_id=p.customer_id
JOIN inventory i
ON r.inventory_id=i.inventory_id
WHERE MONTH(r.rental_date)='05';'''

rented_in_may=pd.read_sql_query(query,engine)
rented_in_may.head()
```

```python
film=pd.merge(left=film,right=rented_in_may,how='outer',left_on='film_id',right_on='rented_in_may')
film.head()
```

```python
film['rented_in_may']=film['rented_in_may'].fillna(0)
film.head()
```

```python
def val_change(x):
    if x!=0:
        return 1
    else:
        return 0
film['rented_in_may']=film['rented_in_may'].apply(val_change)
film.head()
```

```python
#dropping all the unnecessary columns

film=film[['rental_duration','rental_rate','length','replacement_cost','rating','no_of_actors','category','film_inventory','total_rented','rented_in_may']]
film.head()
```

### 4. Analyze extracted features (X) and transform them. You may need to encode some categorical variables, or scale numerical variables.

```python
X=film.drop('rented_in_may',axis=1)
y=film['rented_in_may']
```

```python
X.dtypes #no need to change any dtypes
```

#### Separate X_train and X_test into numerical and categorical (X_train_cat , X_train_num , X_test_cat , X_test_num)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
import numpy as np

X_train_num=X_train.select_dtypes(np.number)
X_test_num=X_test.select_dtypes(np.number)
X_test_num.head()
```

```python
X_train_cat=X_train.select_dtypes(object)
X_test_cat=X_test.select_dtypes(object)
X_test_cat.head()
```

#### Transforming the numerical data using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(X_train_num)
X_train_norm = transformer.transform(X_train_num)
X_train_norm=pd.DataFrame(X_train_norm,index=X_train_num.index, columns=X_train_num.columns)

X_test_norm= transformer.transform(X_test_num)
X_test_norm=pd.DataFrame(X_test_norm,index=X_test_num.index, columns=X_test_num.columns)
```

```python
X_test_norm
```

#### Encoding the categorical data

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_train_cat)
cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)

X_train_cat_encoded= pd.DataFrame(encoder.transform(X_train_cat).toarray(),index=X_train_cat.index,columns=cols)
X_test_cat_encoded = pd.DataFrame(encoder.transform(X_test_cat).toarray(),index=X_test_cat.index,columns=cols)
```

```python
X_train_cat_encoded
```

```python
X_test_cat_encoded
```

```python
X_train_trans=pd.concat([X_train_norm,X_train_cat_encoded],axis=1)
X_test_trans=pd.concat([X_test_norm,X_test_cat_encoded],axis=1)
```

```python
X_train_trans
```

```python
X_train_trans.shape
```

```python
X_test_trans.shape
```

### 5. Create a logistic regression model to predict 'rented_in_may' from the cleaned data.

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression().fit(X_train_trans,y_train)
predictions=model.predict(X_test_trans)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
model.score(X_train_trans,y_train)
```

```python
model.score(X_test_trans,y_test)
```
